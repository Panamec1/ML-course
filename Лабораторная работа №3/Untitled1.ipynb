```python
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap


from math import exp
from matplotlib import pyplot as plt
```

```python
bets = [1, 2, 3, 4, 5]
degs = [2, 3, 4, 5]
consts = [0.05, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]
eps = 10e-5
```

```python
def linear(x, y, extra_arg=None):
    return np.sum(x * y)


def polynomial(x, y, deg):
    #(<x,x'>+1)^d
    return (np.sum(x * y) + 1) ** deg


def gaussian(x, y, beta):
    # exp(-beta*(|| x - x' ||^2))
    z= x - y
    m =  np.sum(np.fromiter(map(lambda x: x ** 2, z), dtype=float))
    return exp(-beta * m)
```

```python
class SVM:
    def __init__(self, features, classes, C, kernel, extra_arg=None):
        self.features = features
        self.targets = classes
        self.amount = len(classes)
        self.C = C
        self.kernel = kernel
        self.extra_arg = extra_arg
        self.alpha = np.zeros(self.amount)
        
        self.b = 0.0
        self.kernel_matrix = self.kernel_matrix()
        self.features_amount = len(features[0])
        self.starter()

    def kernel_matrix(self):
        return np.array([self.kernel(self.features[x], self.features[y], self.extra_arg) for x in range(
            self.amount) for y in range(self.amount)]).reshape((self.amount, self.amount))

    
    def svmFun(self, index):
        return np.sum(self.alpha * self.targets * self.kernel_matrix[index])

    def __calc_constraints(self, x, y):
        if classes[x] == classes[y]:
            lowest = max(0, self.alpha[x] + self.alpha[y] - self.C)
            highest = min(self.alpha[x] + self.alpha[y], self.C)
        else:
            lowest = max(0, self.alpha[y] - self.alpha[x])
            highest = min(self.C - self.alpha[x] + self.alpha[y], self.C)
        return lowest, highest


    def notEq(self, value):
        new_value = value
        while new_value == value:
            new_value = np.random.randint(0, self.amount)
        return new_value
       

    def K(self, obj):
        return np.fromiter(map(lambda x: self.kernel(
            self.features[x], obj, self.extra_arg), range(self.amount)), dtype=float)

    def stater(self):
        iteration   = 0
        has_changed = False
        failed = 0
        
        
        while iteration < 500:
            for i in range(self.amount):
                error_i = self.svmFun(i) - self.targets[i]
                if (self.targets[i] * error_i > eps and self.alpha[u] > 0) or (self.targets[i] * error_i < -eps and self.alpha[i] < self.C):
                    
                    j = notEq(i)
                    
                    error_j = self.__svm_function(j) - self.targets[j]
                    
                    if classes[i] == classes[j]:
                        low = max(0, self.alpha[i] + self.alpha[j] - self.C)
                        high = min(self.alpha[i] + self.alpha[j], self.C)
                    
                    else:
                        low = max(0, self.alpha[j] - self.alpha[i])
                        high = min(self.C - self.alpha[i] + self.alpha[j], self.C)
                    
                    
                    cov = self.kernel_matrix[i, j]
                    x_self = self.kernel_matrix[i, i]
                    y_self = self.kernel_matrix[j, j]
                    
                    eta = 2 * cov - x_self - y_self
                    
                    if low == high or eta >= 0:
                        continue
                        
                        
                    alpha_i_old = self.alpha[i]
                    alpha_j_old =  self.alpha[j]
                    
                    
                    self.alpha[j] -= self.targets[j] * (error_i - error_j) / eta
                    
                    if self.alpha[j] > high:
                        self.alpha[j] = high
                    if self.alpha[j] < low:
                        self.alpha[j] = low
                    
                    diff_j = alpha_j_old - self.alpha[j]
                    if abs(diff_j) < eps:
                        continue
                    self.alpha[i] += self.targets[i] * self.targets[j] * diff_j
                    has_changed = True
            
            iteration += 1
            if has_changed:
                iteration, has_changed = 0, False
                failed += 1
            if failed >= 3 * 500:
                break
        
        for index in range(self.amount):
            if eps < self.alpha[index] < self.C - eps:
                # w0 = <w,x_i> - y_i
                self.b = np.sum(self.alpha * self.targets * self.kernel_matrix[index]) - self.targets[index]
                return
            
        
        supports = list()
        for index in range(self.amount):
            if eps < self.alpha[index]:
                supports.append(np.sum(self.alpha * self.targets * self.kernel_matrix[index]) 
                                - self.targets[index])
        if len(supports) != 0:
            self.b = np.sum(supports) / len(supports)
        else:
            self.b = 0.0

        
    # Постраение классфикации 
    def predict(self, x):
        # L(x) = sign (Sum_i=1_|D| (alpha_i * y_i * K(x,x')) - w0)
        # K(x,x') - функция ядра
        Lx = np.sign(np.sum(self.alpha * self.targets * self.K(x)) - self.b)
        if Lx != 0:
            return Lx
        return np.random.choice([-1, 1])
```

```python
# Validate отвечает за сопоставление предсказания и реального распределния класса
# от общего числа объектов
def validate(svm):
    return np.sum(np.fromiter(map(
        lambda x: svm.predict(features[x]) == svm.targets[x],
        range(svm.amount)), dtype=int)) / svm.amount

def draw_plot(svm, ker_as_str):
    x_min, y_min = np.amin(svm[0].features, 0)
    x_max, y_max = np.amax(svm[0].features, 0)
    x_step, y_step = (x_max - x_min) / 250, (y_max - y_min) / 250
    x_min, x_max = x_min - 3 * x_step, x_max + 3 * x_step
    y_min, y_max = y_min - 3 * y_step, y_max + 3 * y_step
    x_cell, y_cell = np.meshgrid(np.arange(x_min, x_max, x_step), np.arange(y_min, y_max, y_step))
    mesh_dots = np.c_[x_cell.ravel(), y_cell.ravel()]
    c = np.apply_along_axis(lambda x: svm[0].predict(x), 1, mesh_dots)
    c = np.array(c).reshape(x_cell.shape)
    plt.figure(figsize=(10, 10))
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    col = list(map(lambda x: "g" if svm[0].targets[x] == 1 else "b", range(svm[0].amount)))
    plt.pcolormesh(x_cell, y_cell, c, cmap=ListedColormap(['#9090FF', '#90FF90']), shading='nearest')
    plt.scatter(svm[0].features[:, 0], svm[0].features[:, 1], c=col)
    plt.title("Kernel -> "+str(ker_as_str)
              +" C -> "+str(svm[0].C)+" Param -> "+ str(svm[0].extra_arg)+" Accuracy ->"+str(svm[1] * 100))
    plt.show()
```

```python
data = open("chips.csv", "r")
dataset = pd.read_csv(data).values
data.close()
```

```python
features = np.array(dataset[:, :-1])
classes = np.fromiter(map(lambda x: 1 if x == 'P' else -1, dataset[:, len(features[0])]), dtype=int)
linears, polynomials, gaussians = list(), list(), list()

for C in consts:
    print("Const " + str (C))
    print(" -> Linear")
    current = SVM(features, classes, C, linear)
    linears.append((current, validate(current)))
    for degree in degs:
        print(" -> Polynom -> d = " + str(degree))
        current = SVM(features, classes, C, polynomial, extra_arg=degree)
        polynomials.append((current, validate(current)))
    for beta in bets:
        print(" -> Gauss -> b = " + str(beta))
        current = SVM(features, classes, C, gaussian, extra_arg=beta)
        gaussians.append((current, validate(current)))
```
